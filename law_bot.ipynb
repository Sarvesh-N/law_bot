```python
import re

def clean_text(text):
    if not text:
        return ""
    text = re.sub(r"\s+", " ", text)   # remove multiple spaces/newlines
    text = text.strip()
    return text

# Apply cleaning
def preprocess(example):
    return {
        "title": clean_text(example.get("title", "")),
        "section": clean_text(example.get("section", "")),
        "text": clean_text(example.get("text", ""))
    }

ds_clean = dataset.map(preprocess)
```

# Creating Embedings

```python
import pandas as pd

df = pd.read_csv("indian_laws_clean.csv")

# Create a unified column
df["content"] = (
    "Act: " + df["act_title"].astype(str) +
    " | Section: " + df["section"].astype(str) +
    " | Law: " + df["law"].astype(str)
)

print(df.head())
```

```python
df.columns
```

```python
df = df.drop(columns = 'title')
df = df.drop(columns = 'text')
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

model = SentenceTransformer("all-MiniLM-L6-v2")

# Encode using 'content' column
embeddings = model.encode(df["content"].tolist(), show_progress_bar=True)
embeddings = np.array(embeddings).astype("float32")

# Create FAISS index
index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(embeddings)
```

```python
query = "punishment for drunk and driving"
query_embedding = model.encode([query])

distances, indices = index.search(np.array(query_embedding).astype("float32"), 5)

for i in indices[0]:
    print(df.iloc[i]["content"])
```

```python
import faiss
import pickle

# Save FAISS index
faiss.write_index(index, "law_index.faiss")

# Save dataframe mapping (so we know which row maps to which embedding)
with open("law_mapping.pkl", "wb") as f:
    pickle.dump(df, f)
```
